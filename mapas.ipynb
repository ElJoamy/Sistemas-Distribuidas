```python
import os
import graphviz
#descargar graphviz de https://graphviz.org/download/
# instalar la ultima version de windows .exe

file_name = 'mapa_conceptual'

if os.path.exists(file_name + '.pdf'):
    i = 1
    while os.path.exists(file_name + str(i) + '.pdf'):
        i += 1
    file_name += str(i)

dot = graphviz.Digraph()

# nodos
dot.node('A', 'Sistemas Distribuidos')
dot.node('B', 'RAID: Conjunto Redundante de Discos Independientes')
dot.node('C', 'LVM: Administrador de Volúmenes Lógicos')
dot.node('D', 'Modelo de Interacción en Sistemas Distribuidos')
dot.node('E', 'Modelo de Fallos en Sistemas Distribuidos')
dot.node('F', 'Llaves Asimétricas')
dot.node('G', 'Replicación en Sistemas Distribuidos')
dot.node('H', 'Exclusión Mutua en Sistemas Distribuidos')
dot.node('I', 'Relojes Lógicos y Físicos')
dot.node('J', 'Transacciones en Sistemas Distribuidos')
dot.node('K', 'DNS: Sistema de Nombres de Dominio')
dot.node('L', 'NGINX: Servidor Web y Proxy Inverso')
dot.node('M', 'Docker: Plataforma de Contenedores')
dot.node('N', 'Proxmox VE: Plataforma de Virtualización')
dot.node('O', 'Zabbix: Sistema de Monitorización de Redes y Servidores')
dot.node('P', 'Veeam Backup & Replication Community Edition')
dot.node('Q', 'Conclusión')

# relaciones
dot.edge('A', 'B')
dot.edge('A', 'C')
dot.edge('A', 'D')
dot.edge('A', 'E')
dot.edge('A', 'F')
dot.edge('A', 'G')
dot.edge('A', 'H')
dot.edge('A', 'I')
dot.edge('A', 'J')
dot.edge('A', 'K')
dot.edge('A', 'L')
dot.edge('A', 'M')
dot.edge('A', 'N')
dot.edge('A', 'O')
dot.edge('A', 'P')
dot.edge('A', 'Q')

# subtemas
dot.node('B1', 'Introducción')
dot.node('B2', 'Sistemas Distribuidos')
dot.node('B3', 'RAID (Redundant Array of Independent Disks)')
dot.node('C1', 'LVM (Logical Volume Manager)')
dot.node('D1', 'Modelo de Interacción en Sistemas Distribuidos')
dot.node('E1', 'Modelo de Fallos en Sistemas Distribuidos')
dot.node('F1', 'Llaves Asimétricas')
dot.node('G1', 'Replicación en Sistemas Distribuidos')
dot.node('H1', 'Exclusión Mutua en Sistemas Distribuidos')
dot.node('I1', 'Relojes Lógicos y Físicos')
dot.node('J1', 'Transacciones en Sistemas Distribuidos')
dot.node('K1', 'DNS (Domain Name System)')
dot.node('L1', 'NGINX: Servidor Web y Proxy Inverso')
dot.node('M1', 'Docker: Plataforma de Contenedores')
dot.node('N1', 'Proxmox VE: Plataforma de Virtualización')
dot.node('O1', 'Zabbix: Sistema de Monitorización de Redes y Servidores')
dot.node('P1', 'Veeam Backup & Replication Community Edition')
dot.node('Q1', 'Conclusión')

# relaciones con subtemas
dot.edge('B', 'B1')
dot.edge('B', 'B2')
dot.edge('B', 'B3')
dot.edge('C', 'C1')
dot.edge('D', 'D1')
dot.edge('E', 'E1')
dot.edge('F', 'F1')
dot.edge('G', 'G1')
dot.edge('H', 'H1')
dot.edge('I', 'I1')
dot.edge('J', 'J1')
dot.edge('K', 'K1')
dot.edge('L', 'L1')
dot.edge('M', 'M1')
dot.edge('N', 'N1')
dot.edge('O', 'O1')
dot.edge('P', 'P1')
dot.edge('Q', 'Q1')

# definiciones
dot.node('B1.1', 'RAID es una tecnología que combina múltiples\n' 
         'discos duros independientes en un único sistema de\n'
         'almacenamiento para mejorar la redundancia y/o el rendimiento.')
dot.node('B3.1', 'El RAID consiste en dividir y distribuir los\n' 
         'datos en varios discos para ofrecer mayor tolerancia\n' 
         'a fallos y rendimiento mejorado.')
dot.node('C1.1', 'LVM es una herramienta de administración de\n'
         'volúmenes lógicos que permite combinar y administrar varios\n'
         'dispositivos de almacenamiento en un único espacio de almacenamiento\n'
         'lógico.')
dot.node('D1.1', 'El modelo de interacción en sistemas distribuidos\n'
         'se refiere a cómo los distintos componentes y nodos de un sistema\n'
         'distribuido se comunican y colaboran entre sí.')
dot.node('E1.1', 'El modelo de fallos en sistemas distribuidos se ocupa\n'
         'de cómo los sistemas distribuidos pueden manejar y recuperarse\n'
         'de los fallos que pueden ocurrir en los componentes individuales.')
dot.node('F1.1', 'Las llaves asimétricas son un método criptográfico que\n'
         'utiliza un par de claves: una clave pública para cifrar y una\n'
         'clave privada para descifrar. Se utiliza en la criptografía de\n'
         'clave pública.')
dot.node('G1.1', 'La replicación en sistemas distribuidos consiste en\n'
         'mantener copias idénticas de los datos en diferentes nodos para\n'
         'garantizar la disponibilidad y la tolerancia a fallos.')
dot.node('H1.1', 'La exclusión mutua en sistemas distribuidos se refiere\n'
         'a cómo se evita que múltiples procesos accedan simultáneamente\n'
         'a un recurso compartido para evitar inconsistencias y conflictos.')
dot.node('I1.1', 'Los relojes lógicos y físicos son mecanismos utilizados\n'
         'en sistemas distribuidos para sincronizar eventos y secuenciar\n'
         'operaciones en diferentes nodos.')
dot.node('J1.1', 'Las transacciones en sistemas distribuidos se utilizan\n'
         'para asegurar la integridad y consistencia de las operaciones\n'
         'realizadas en un conjunto de nodos, garantizando la atomicidad,\n'
         'la consistencia, el aislamiento y la durabilidad (ACID).')
dot.node('K1.1', 'DNS es un sistema de nombres de dominio que asocia\n'
         'nombres de dominio legibles para los humanos con direcciones\n'
         'IP numéricas utilizadas por las computadoras para comunicarse en Internet.')
dot.node('L1.1', 'NGINX es un servidor web y proxy inverso de alto\n'
         'rendimiento que se utiliza para balanceo de carga, caching,\n'
         'manejo de solicitudes HTTP y HTTPS, y como servidor de archivos estáticos.')
dot.node('M1.1', 'Docker es una plataforma de contenedores que permite\n'
         'empaquetar y distribuir aplicaciones junto con todas sus\n'
         'dependencias en entornos aislados, facilitando la implementación\n'
         'y el despliegue de aplicaciones.')
dot.node('N1.1', 'Proxmox VE es una plataforma de virtualización de\n'
         'servidores de código abierto que combina virtualización basada\n'
         'en contenedores y virtualización basada en hipervisores para\n'
         'gestionar y desplegar máquinas virtuales y contenedores.')
dot.node('O1.1', 'Zabbix es un sistema de monitorización de redes y servidores\n'
         'que supervisa y registra el rendimiento de los dispositivos\n'
         'y servicios de red, generando alertas y notificaciones en caso\n'
         'de eventos o condiciones anormales.')
dot.node('P1.1', 'Veeam Backup & Replication Community Edition es una\n'
         'solución de copia de seguridad y recuperación de desastres que\n'
         'proporciona protección de datos y replicación para entornos virtuales\n'
         'y físicos.')
dot.node('Q1.1', 'En conclusión, los sistemas distribuidos implican la\n'
         'interconexión de múltiples nodos y componentes para proporcionar\n'
         'redundancia, escalabilidad, tolerancia a fallos y rendimiento mejorado.\n'
         'Los temas discutidos en el mapa conceptual representan conceptos clave\n'
         'en los sistemas distribuidos, como RAID, LVM, modelos de interacción y\n'
         'fallos, replicación, exclusión mutua, relojes, transacciones, DNS, NGINX,\n'
         'Docker, Proxmox, Zabbix y Veeam Backup & Replication.')

# elaciones con definiciones
dot.edge('B1.1', 'B1')
dot.edge('B3.1', 'B3')
dot.edge('C1.1', 'C1')
dot.edge('D1.1', 'D1')
dot.edge('E1.1', 'E1')
dot.edge('F1.1', 'F1')
dot.edge('G1.1', 'G1')
dot.edge('H1.1', 'H1')
dot.edge('I1.1', 'I1')
dot.edge('J1.1', 'J1')
dot.edge('K1.1', 'K1')
dot.edge('L1.1', 'L1')
dot.edge('M1.1', 'M1')
dot.edge('N1.1', 'N1')
dot.edge('O1.1', 'O1')
dot.edge('P1.1', 'P1')
dot.edge('Q1.1', 'Q1')

dot.graph_attr['dpi'] = '100'   
dot.graph_attr['fontsize'] = '30'
dot.graph_attr['fontname'] = 'Times New Roman'
dot.graph_attr['bgcolor'] = '#E8E8E8'
dot.attr(label='Mapa conceptual de sistemas distribuidos', labelloc='t', fontname='Arial', fontsize='40')
dot.attr(labelloc='t')

# Guardar el gráfico en un archivo
dot.format = 'pdf'
dot.render(file_name, view=True)
```
